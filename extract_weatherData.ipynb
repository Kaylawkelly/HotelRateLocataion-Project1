```python
# Dependencies
import pandas as pd
import numpy as np
import json
import requests
from datetime import datetime
from config import api_key
```

```python
cities_pd = pd.read_csv("./Resources/Airport_Output.csv")
cities_pd=cities_pd.head()
print(cities_pd)
```

```python
city_name=[]
lat=[]
lng=[]
dateTime=[]
date=[]
time=[]
sunrise=[]
sunset=[]
pressure=[]
humidity=[]
wind_speed=[]
low_temp=[]
high_temp=[]
descriptor=[]
```

```python
for index, row in cities_pd.iterrows():
    cur_lat = row['Lat']
    cur_lng = row['Lng']
    # Save config information
    url="https://api.openweathermap.org/data/2.5/onecall?units=imperial&lat=" + str(cur_lat) + "&lon=" + str(cur_lng) + "&"
    # Build query URL
    query_url = url + "appid=" + api_key
    #print(query_url)
    # Get weather data
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    num_days=len(weather_json['daily'])
    for x in range(num_days):
        # Get the temperature from each day of the response
        #print(x)
        city_name.append(row['City'])
        lat.append(cur_lat)
        lng.append(cur_lng)
        dt_object=datetime.fromtimestamp(weather_json['daily'][x]['dt'])
        timestampStr = dt_object.strftime("%d-%b-%Y (%H:%M:%S.%f)")
        time_list=timestampStr.split()
        dateTime.append(time_list[0])
        dt_object=datetime.fromtimestamp(weather_json['daily'][x]['sunrise'])
        timestampStr = dt_object.strftime("%d-%b-%Y (%H:%M:%S.%f)")
        time_list=timestampStr.split()
        sunrise.append(time_list[1])
        dt_object=datetime.fromtimestamp(weather_json['daily'][x]['sunset'])
        timestampStr = dt_object.strftime("%d-%b-%Y (%H:%M:%S.%f)")
        time_list=timestampStr.split()
        sunset.append(time_list[1])
        pressure.append(weather_json['daily'][x]['pressure'])
        humidity.append(weather_json['daily'][x]['humidity'])
        wind_speed.append(weather_json['daily'][x]['wind_speed'])
        low_temp.append(weather_json['daily'][x]['temp']['min'])
        high_temp.append(weather_json['daily'][x]['temp']['max'])
        descriptor.append(weather_json['daily'][x]['weather'][0]['main'])
```

```python
weather_dict = {
    "city": city_name,
    "lat": lat,
    "lng": lng,
    "dateTime": dateTime,
    "sunrise": sunrise,
    "sunset": sunset,
    "pressure": pressure,
    "humidity": humidity,
    "wind_speed (MPH)": wind_speed,
    "low_temp (degF)": low_temp,
    "high_temp (degF)": high_temp,
    "descriptor": descriptor,
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head(20)
```

```python
weather_data.to_csv("Resources/city_weather_data.csv", index=False, header=True)
```
